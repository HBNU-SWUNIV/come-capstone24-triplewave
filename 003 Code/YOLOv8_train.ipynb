```python
from ultralytics import YOLO
import wandb

# WandB 로그인
wandb.login(key='38136729c5b8fd06aa7b7b7acc566f1f98a47524')

# 이미 학습된 모델 로드
path = '/kaggle/working/runs/detect/train5/weights/best.pt'
model = YOLO(path)

# 새로운 데이터셋 경로
new_dataset = "/kaggle/input/signboard/school_project.v4i.yolov8/data.yaml"

# 모델의 백본 부분을 고정하여 분류기만 학습하도록 설정
for param in model.model.backbone.parameters():
    param.requires_grad = False

# 모델 훈련 (분류기만 전이 학습)
model.train(
    data=new_dataset,
    epochs=10,
    device=[0, 1],  # 두 개의 GPU 사용
    patience=3,  # 조기 종료 설정
    label_smoothing=0.2  # 필요에 따라 추가 파라미터 설정
)
```
